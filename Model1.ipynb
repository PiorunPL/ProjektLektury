# Model 1
Poniższy model w celu wyznaczenia najbardziej prawdopodobnego autora wykorzystuje następującą konfigurację:
   -  Worki słów tworzone są dla każdego autora (a nie dla pojedynczej lektury)
   -  Worki słów podlegają lematyzacji
   -  Z worków słów usuwane są słowa stopu
    
Wykorzystywane metody znajdują się w pliku Measures.ipynb oraz WordBag.ipynb

    

### Importy i przygotowanie


```python
import ipynb.fs.full.WordBag as WordBag
import ipynb.fs.full.Measures as Measures
import operator
```

```python
words_bags = WordBag.create_bags_of_words_from_authors(False, False, True)
```

### Test 1
Na początku przetestujemy nasz model, na krótkim fragmencie "Pana Tadeusza" Adama Mickiewicza

```python
fragment = """    Litwo! Ojczyzno moja! ty jesteś jak zdrowie:
Ile cię trzeba cenić, ten tylko się dowie,
Kto cię stracił. Dziś piękność twą w całej ozdobie
Widzę i opisuję, bo tęsknię po tobie."""


possible_authors = Measures.find_possible_authors_percents_v3(fragment, words_bags, False, True)
sorted_dict = dict( sorted(possible_authors.items(), key=operator.itemgetter(1),reverse=True))

keys = sorted_dict.keys()

for key in keys:
    print(key + ': '+ str(sorted_dict[key]) + ' %')

#print(sorted_dict)

```

### Test 5
W tej części dokonamy testu na 108 fragmentach, by sprawdzić w ilu przypadkach nasz program zwróci poprawnego autora. Dodatkowo wykonamy pomiar, potrzebny do ogólnego porównania, który model jest najlepszy. Nie chcemy oceniać modelu tylko i wyłącznie po ilości poprawnie wskazanych najbardziej prawdopodobnych autorów, ponieważ, może się zdażyć tak, że model wskaże innego autora nieznaczną przewagą, natomiast ogólnie zawsze wysoko stawiał poprawnego autora (chociaż nie zawsze on wygrywał). By tego dokonać zsumujemy procenty przyznawane przez model poprawnym autorom. W części podsumowującej modele dokonamy analizy.

```python
number_of_fragments = 108
number_of_correctly_chosen_authors = 0
sum_of_percents = 0

print('Aktualnie pracuję nad: ')

for i in range(1, number_of_fragments+1):
    file_name = 'Fragments/fragment-' + str(i) + '.txt'
    possible_authors_percents = Measures.find_possible_authors_percents_v2(file_name, words_bags, True, False, True)
    with open(file_name,'r', encoding='utf-8') as file:
        print(file_name)
        author = WordBag.get_author(file)
    sum_of_percents += possible_authors_percents[author]
    all_authors = words_bags.keys()
    #sorted_dict = dict( sorted(possible_authors_percents.items(), key=operator.itemgetter(1),reverse=True))
    #print(sorted_dict)
    most_possible_author = max(possible_authors_percents, key=possible_authors_percents.get)
    if most_possible_author == author:
        number_of_correctly_chosen_authors+=1

print('Procentowa skuteczność danego modelu: ' + str(number_of_correctly_chosen_authors/number_of_fragments*100) + ' %')
print('Suma procent poprawnych autorów: ' + str(sum_of_percents))

```
