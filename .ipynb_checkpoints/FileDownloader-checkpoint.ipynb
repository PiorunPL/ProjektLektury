Powyższe dzieła zostały wybrane w taki sposób, aby były zarówno przypadki podobnych do siebie jak i bardzo wyróżniające się na tle innych dzieł. Są to pozycje z różnych epok i różnych rodzajów.  

```python
import json
import requests
from urllib.parse import urlparse

url = 'https://wolnelektury.pl/api/books/?'
r = requests.get(url, allow_redirects=True)
data =r.json()

for i in data:
    href = i['href']
    rBook = requests.get(href, allow_redirects=True)
    #fileTitle
    if href.find('/'):
        a = urlparse(href)
        table = href.split('/')
        fileTitle = table[len(table) - 2]
    dataBook = rBook.json()
    title = dataBook['title']
    txt = dataBook['txt']
    if txt:
        rText = requests.get(txt, allow_redirects=True)
        filename = 'Books/' + fileTitle + '.txt'
        open(filename, 'wb').write(rText.content)
        print(filename)
    else:
        print('\t\t'+'title')
        
        
        
```
